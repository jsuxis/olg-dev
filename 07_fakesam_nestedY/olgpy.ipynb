```python
import gdxpds
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from pylab import rcParams

rcParams['figure.figsize'] = 15, 5
gdx_file = 'C:/Users/Manu/Documents/olg/03_7sec_sigAge/results_dem1.gdx'
gdx_file2 = 'C:/Users/Manu/Documents/olg/03_7sec_sigAge/results_dem2.gdx'
gdx_file3 = 'C:/Users/Manu/Documents/olg/03_7sec_sigAge/results_prod.gdx'
dataframes = gdxpds.to_dataframes(gdx_file)
dataframes2 = gdxpds.to_dataframes(gdx_file2)
dataframes3 = gdxpds.to_dataframes(gdx_file3)
```

```python
Y = dataframes['Y']
Y['Level2'] = dataframes2['Y']['Level']
Y['Level3'] = dataframes3['Y']['Level']


macrovariables = ['K','I','Bond','Rent','Rint']
macro = {}
for v in macrovariables:
    macro[v] = dataframes[v]
    macro[v]['Level2'] = dataframes2[v]['Level']
    macro[v]['Level3'] = dataframes3[v]['Level']
```

```python
Y1 = Y[Y['s'] == 's1']
Y2 = Y[Y['s'] == 's2']
Y3 = Y[Y['s'] == 's3']
Y4 = Y[Y['s'] == 's4']
Y5 = Y[Y['s'] == 's5']
Y6 = Y[Y['s'] == 's6']
Y7 = Y[Y['s'] == 's7']
#Y8 = Y[Y['s'] == 's8']
    
Yi = [Y1,Y2,Y3,Y4,Y5,Y6,Y7]
for i in Yi:
    fig = plt.figure()
    ax  = fig.add_subplot(1, 1, 1)
    ax.set_ylabel(i.s.iloc[0], color='black') 
    ax.tick_params(axis='x', colors='black')
    ax.tick_params(axis='y', colors='black')
    plt.plot(i.t,i.Level, label='Baby Boom') 
    plt.plot(i.t, i.Level2, label='Baby Boom and Bust')
    plt.plot(i.t, i.Level3, label='Productivity Sector 1 increases')
    ax.axvline('t4', color='k', linestyle='--')
    ax.axvline('t10', color='k', linestyle='--')
    plt.legend(loc='upper right')
    plt.savefig("C:/Users/Manu/Documents/olg/03_7sec_sigAge/plotY"+str(i.s.iloc[0])+".png")
```

```python
for v in macrovariables:
    fig = plt.figure()
    ax  = fig.add_subplot(1, 1, 1)
    ax.set_ylabel(v, color='black') 
    ax.tick_params(axis='x', colors='black')
    ax.tick_params(axis='y', colors='black')
    
    plt.plot(macro[v].t,macro[v].Level, label='Baby Boom') 
    plt.plot(macro[v].t,macro[v].Level2, label='Baby Boom and Bust')
    plt.plot(macro[v].t,macro[v].Level3, label='Productivity Sector 1 increases')
    ax.axvline('t4', color='k', linestyle='--')
    ax.axvline('t10', color='k', linestyle='--')
    plt.legend(loc='upper right')
    plt.savefig("C:/Users/Manu/Documents/olg/03_7sec_sigAge/plot"+str(v)+".png")
```
